```python
#NASA Mars news

#Website info
url = "https://redplanetscience.com/"
browser.visit(url)

#read webpage
html = browser.html
soup = bs(html, 'html.parser')
print(soup.prettify())
```

```python
#look for article titles & paragraph text
result = soup.find('div',class_='list_text')

print(result)
```

```python
#get title & store in variable
news_title = result.find(class_='content_title').text.strip()
print(news_title)
```

```python
#get text  & store in variable
news_article = result.find(class_='article_teaser_body').text.strip()
print(news_article)
```

```python
#Mars Featured Image

#Website info
url = "https://spaceimages-mars.com/"
browser.visit(url)

#read webpage
html = browser.html
soup = bs(html, 'html.parser')
print(soup.prettify())
```

```python
#look for image & find the url
result = soup.find('div',class_='floating_text_area')

featured_image = result.find('a')['href']
featured_image_url = ('https://spaceimages-mars.com/' + featured_image)
#print(result)
print(featured_image_url)
```
