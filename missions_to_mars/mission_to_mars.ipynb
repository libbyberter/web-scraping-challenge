```python
# Dependencies
import os
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#NASA Mars news

#Website info
url = "https://redplanetscience.com/"
browser.visit(url)

#read webpage
html = browser.html
soup = bs(html, 'html.parser')
print(soup.prettify())
```

```python
#look for article titles & paragraph text
result = soup.find('div',class_='list_text')

print(result)
```

```python
#get title & store in variable
news_title = result.find(class_='content_title').text.strip()
print(news_title)
```

```python
#get text  & store in variable
news_article = result.find(class_='article_teaser_body').text.strip()
print(news_article)
```

```python
#Mars Featured Image

#Website info
url2 = "https://spaceimages-mars.com/"
browser.visit(url2)

#read webpage
html = browser.html
soup = bs(html, 'html.parser')
print(soup.prettify())
```

```python
#Hemispheres

#Website info
hemispheres_url = "https://marshemispheres.com/"
browser.visit(hemispheres_url)

#read webpage
html = browser.html
soup = bs(html, 'html.parser')
print(soup.prettify())
```

```python
#list for hemispheres
hemi_list=[]

#dictionary for hemisphere data
hemi_dict_list=[]
```

```python
#get links for the hemispheres

hemispheres = soup.find_all('div', class_='description')
#print(hemispheres)

for hemisphere in hemispheres:
    a = hemisphere.find('a')
    href = a['href']
    link = (hemispheres_url + href)
    hemi_list.append(link)
    
print(hemi_list)    
    
```

```python
for i in hemi_list:
    
    #Website info
    hemi_url = i
    browser.visit(hemi_url)

    #read webpage
    html = browser.html
    soup = bs(html, 'html.parser')
    #print(soup.prettify())

    #image
    result = soup.find('div', id="wide-image", class_="wide-image-wrapper")
    result_image = result.find(class_='wide-image')['src']
    result_image_url = (hemispheres_url + result_image)
    print(result_image_url)

    #title
    result_title = soup.find('div', class_="cover")
    result_title_text = result_title.find('h2', class_='title').text.strip()
    print(result_title_text)
    
    print('---------------')
    
    hemi_dict = ({"title": result_title_text, "img_url": result_image_url})
    
    hemi_dict_list.append(hemi_dict)
    
print(hemi_dict_list)
```
