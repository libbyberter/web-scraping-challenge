```python
html = browser.html
soup = bs(html, 'html.parser')
print(soup.prettify())
```

```python
#look for article titles
result = soup.find('div',class_='list_text')

#store title in variable
news_title = result.find(class_='content_title').text.strip()
print(news_title)
```
